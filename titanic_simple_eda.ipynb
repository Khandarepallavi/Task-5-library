```

# 1. Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 2. Load dataset
titanic = sns.load_dataset('titanic')

# 3. Basic Information
print(titanic.info())
print(titanic.describe())
print(titanic['sex'].value_counts())
print(titanic['class'].value_counts())
print(titanic['embarked'].value_counts())

# 4. Visualizations

# Pairplot
sns.pairplot(titanic.dropna(), hue='survived')
plt.show()

# Heatmap
sns.heatmap(titanic.corr(), annot=True, cmap='coolwarm')
plt.show()

# Histogram
titanic['age'].hist(bins=30, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Boxplot
sns.boxplot(x='class', y='age', data=titanic)
plt.title('Age by Passenger Class')
plt.show()

# Scatterplot
sns.scatterplot(x='age', y='fare', hue='survived', data=titanic)
plt.title('Fare vs Age (colored by Survival)')
plt.show()

# 5. Observations
print("""
Observations:
- Most passengers were young adults (20-40 years old).
- Higher class (1st class) passengers paid higher fares and were older.
- Survival rate was higher for females and first-class passengers.
- Strong correlations between 'fare', 'class', and 'survived'.
""")

# 6. Summary
print("""
Summary:
EDA shows that gender, passenger class, and fare significantly influenced survival rates.
Younger and first-class passengers had better survival chances.
Further cleaning and modeling can improve understanding.
""")
```
